```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt


def plotter(n):
    G=nx.hypercube_graph(n)
    nx.draw(G)
    plt.show()

from ipywidgets import interactive
from IPython.display import display
w = interactive(plotter, n=(1,6))
display(w)
```

```python



def plotter(n,m):
    w = np.random.random((n,m))
    G = nx.DiGraph()

    #add the nodes and edges to the graph
    for i in range(n):
        for j in range(m):
            G.add_edge(i,j+n,weight = w[i][j])

    #set the positions as a bipartide graph layout
    pos = dict()
    pos.update( (i, (1, i*2)) for i in range(n) )    
    pos.update( (j+n, (10, j)) for j in range(m) )
    
    nx.draw(G,pos)
    labels = nx.get_edge_attributes(G,'weight')
    nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
    plt.show()
    
from ipywidgets import interactive
from IPython.display import display
w = interactive(plotter, n=(2,6),m = (1,5))
display(w)
```
