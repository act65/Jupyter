```python
import tensorflow as tf
sess = tf.InteractiveSession()
init = tf.initialize_all_variables()
```

```python
with tf.name_scope("lolz") as scope:
    a = tf.placeholder(tf.float32,[1], name="a")
    b = tf.placeholder(tf.float32,[1], name="b")
    c = a + b
    tf.scalar_summary(["c"], c)
```

```python
with tf.Session() as sess:
    writer = tf.train.SummaryWriter("/tmp/tf1boardcheck", sess.graph)
    merged = tf.merge_all_summaries()

    sess.run(init)
    for i in range(10):
        summary,C = sess.run([merged,c],feed_dict={a:[i**2],b:[2]})
        writer.add_summary(summary,i)
```
