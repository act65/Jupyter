```python
import tensorflow as tf
weights = tf.Variable(tf.zeros([784, 200]),name="weights")
biases = tf.Variable(tf.zeros([200,1]), name="biases")
saver = tf.train.Saver()

with tf.Session() as sess:
    saver.restore(sess, "/tmp/model.ckpt") #ok cool, so this over writes weights and biases defined before
    print("Model restored.")
    ans = sess.run(tf.matmul(weights,biases))
    print(ans)
```
