```python
import gym
import numpy as np
env = gym.make('CartPole-v0')
n = 20
m = 1000
actions = np.zeros((n,m))
for e in range(n):
    total = 0
    reward = 0
    observation = env.reset()
    for t in range(m):
        
        #derivative control
        if observation[3] >0:
            action = 1
        elif observation[3] <0:
            action = 0
        else:
            action = None
            
        #observation = x, x_dot, theta, theta_dot
        if action != None:
            actions[e,t] = action
            observation, reward, done, info = env.step(action)
        total += reward
        if done:
            print("Episode {} finished after {} timesteps. Reward = {}".format(e,t,round(total,3)))
            print("Position {}, angle {}\n".format(observation[0],observation[2]))
            break
```
