```python
import matplotlib
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (15, 10)
import matplotlib.pyplot as plt

import numpy as np
import cmath
```

$$
e^{ix} = \cos x +  i\sin x \\
e^{a + bi} = e^a \cdot e^{bi} = e^a(\cos b +  i\sin b)$$

```python
def Complex(A,B):
    x = A.shape
    y = B.shape
    if x == y:
        n = A.size
        A = A.flatten()
        B = B.flatten()
        C = np.zeros(n,dtype = 'complex128')
        for i in range(n):
            C[i] = complex(A[i],B[i])
        return C.reshape(x)
```

```python
n = 1000
x = np.linspace(0,20,n)
y = Complex(np.zeros(n),x)

plt.plot(x,np.real(np.exp(y)),label = 'real')
plt.plot(x,np.imag(np.exp(y)),label = 'imag')
plt.legend()
```

```python
x = np.zeros(n)
y = np.linspace(0,20,n)
z = np.exp(Complex(x,y))

t = np.linspace(0,10,n)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(t,np.imag(z),np.real(z))
ax.set_xlabel('t')
ax.set_ylabel('Imag')
ax.set_zlabel('Real')
```

```python
### Vary the base
x = np.zeros(n)
y = np.linspace(0,20,n)
t = np.linspace(0,200,n)
z = t**(Complex(x,y))

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(t,np.imag(z),np.real(z))
ax.set_xlabel('t')
ax.set_ylabel('Imag')
ax.set_zlabel('Real')
```

```python
### Allow real coefficients in the exponent
x = np.linspace(0,5,n)
y = np.linspace(0,20,n)
z = np.exp(Complex(x,y))

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x,np.imag(z),np.real(z))
ax.set_xlabel('x')
ax.set_ylabel('Imag')
ax.set_zlabel('Real')
```

```python
### Invert base and real exponent
x = np.linspace(0,10,n)
y = np.linspace(0,10,n)
t = np.linspace(10,0.1,n)
z = t**(Complex(x,y))

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(t,np.imag(z),np.real(z))
ax.set_xlabel('t')
ax.set_ylabel('Imag')
ax.set_zlabel('Real')
```

```python
### Complex Sigmoid 
x = np.zeros(n)
y = np.linspace(-15,15,n)
z = np.exp(Complex(x,50/(1+np.exp(-y))))

t = np.linspace(0,10,n)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(t,np.imag(z),np.real(z))
ax.set_xlabel('t')
ax.set_ylabel('Imag')
ax.set_zlabel('Real')
```
