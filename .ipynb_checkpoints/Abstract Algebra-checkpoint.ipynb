### Equivalence relations

* Reflective $x\sim x$
* Symmetric $x\sim y \implies y \sim x$
* Transitive $x\sim y, y\sim z \implies x \sim z$

### Equivalence class
* For $a\in X, [a] = \{b\;|\; a\sim b:b\in X\}$
