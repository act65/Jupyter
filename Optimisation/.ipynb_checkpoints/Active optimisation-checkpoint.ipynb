# Non-convex

But it isnt that simple in in real life... Non-convexity





````python
n = 100
x = np.linspace(-10,10,100)

def g(x):
    return (x-3)*(x-8.0)*(x+2.0)*(x+0.1)/25

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, g(x))
ax1.scatter(6.5, -13,c='k')
ax1.set_title('Easy optimisation')
ax1.set_xlim([-10,10])
ax1.set_ylim([-20,120])

ax2.scatter(5, 25,c='r')
ax2.scatter(0, 0,c='k')
ax2.plot(x, 10*x-25,c='g')
ax2.plot(x, g(x),'b--')
ax2.set_title('Gradient descent')
ax2.set_xlim([-10,10])
ax2.set_ylim([-20,120])
````
