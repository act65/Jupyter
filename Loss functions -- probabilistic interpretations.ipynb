```julia
#ML
ML(X) = -log(X)

#CE
CE(X,Y) = -X*log(Y)

#KL -- is related to bergman divergence?
KL(X,Y) = -X*log(X/Y)

#JS
function JS(X,Y,g=0.5)
    Z = g*X + (1-g)*Y #or is it not symmetric? are these in the right order?
    return g*KL(X,Z) + (1-g)*KL(Y,Z)
end

```
